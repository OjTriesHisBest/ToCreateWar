```python
import random

values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':11,
         'Queen':12, 'King':13, 'Ace':14}

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
```

```python
class Card: 
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
'''Deck 

Create new deck: 
    Create all 52 cards
    Holds list of card objects 
    Shuffle through method call
    Deal cards from Deck object
    pop method
    
Deck class will hold a list of Card objects. It returns Card object 
class instances, not just python data types.'''

class Deck: 
    
    def __init__(self): # doesn't need user input
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                created_card = Card(suit, rank) #for every suit and rank, you'll be putting a suit and a card together. This follows the logic from the Card class
                self.all_cards.append(created_card) #this will add to the empty list
    
    def shuffle(self):
        random.shuffle(self.all_cards) #this doesn't return anything.
        
    def deal_one(self): 
        return self.all_cards.pop() # takes list, removes one and returns it
```

```python
'''Player class 
    Holds current list of cards 
    PLayers should be able to remove or add cards from their hands
    
    Should have a top and bottom. Draw from top and add to bottom'''

class Player:
    #.extend method just adds multple lists together at the end (list method)
    #.append would give me a nested list (like glued together cards)
    
    def __init__(self,name):
        self.name = name
        self.all_cards = []
        
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def add_cards(self, new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards) #multiple cards
        else:
            self.all_cards.append(new_cards) #single card
    
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.'
    
    
```

```python
round_num = 0

#if this cell is run, the whole game starts again.
player_one = Player("One")
player_two = Player("Two")

new_deck = Deck()
new_deck.shuffle()

new_deck

for x in range(26):
    player_one.add_cards(new_deck.deal_one())
    player_two.add_cards(new_deck.deal_one())

# while game_on 
game_on = True

while game_on:
    
    round_num += 1
    #print(f"Round {round_num}")
    
    if len(player_one.all_cards) == 0:
        print('Player One has lost! Player One wins')
        game_on = False 
        break
        
    if len(player_two.all_cards) == 0:
        print('Player Two has lost! Player One wins')
        game_on = False 
        break
    
    # start a new round 
    player_one_cards = [] #we will then append to this list
    player_one_cards.append(player_one.remove_one())
    
    player_two_cards = []
    player_two_cards.append(player_two.remove_one())

    #while at_war
    # 3 situations - 1 > 2, 2 > 3, 1 == 2
    # if there is a tie, people will draw 5 cards. 
    # if they don't have 5 cards to draw, they'd lose.
    
    at_war = True
    
    while at_war:
        
        if player_one_cards[-1].value > player_two_cards[-1].value:
            
            player_one.add_cards(player_one_cards)
            player_one.add_cards(player_two_cards)
            
            at_war = False 
            
        elif player_two_cards[-1].value > player_one_cards[-1].value:
            
            player_two.add_cards(player_two_cards)
            player_two.add_cards(player_one_cards)
            
            at_war = False 
        
        else: 
            print("We're at war!")
            
            if len(player_one.all_cards) < 5:
                print("Player one cannot declare war. Player one wins")
                game_one = False
                break
            elif len(player_two.all_cards) < 5:
                print("Player two cannot declare war. Player one wins")
                game_one = False
                break
            else:
                for num in range(5):
                    player_one_cards.append(player_one.remove_one())
                    player_two_cards.append(player_two.remove_one())
                print(f"Round {round_num}")
```
